```python
from matplotlib import pyplot as plt
import numpy as np

# matplotlib is the main engine/package for plotting, modelled after matlab
## 1. object oriented way (gives you fine control) 2. API
```

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

```python

fig, ax = plt.subplots(nrows=1, ncols=1) # Creating a figure
ax.plot(x, cos_x, label='cos x') # Creating axis
ax.plot(x, sin_x, label='sin x')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2) # Creating a figure
ax[0].plot(x, cos_x, label='cos x', color='red') # Creating axis
ax[0].legend()
ax[0].grid(alpha=.25)
ax[1].plot(x, sin_x, label='sin x')
ax[1].legend()
ax[1].grid(alpha=.25)
# Have control over the axes, easier to change stuff
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, cos_x, label='cos x', color='red')
ax[1].plot(x, sin_x, label='sin x')
ax[0].set_xlabel('Hahaha')


for axis in ax:
    axis.legend()
    axis.grid()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10,3))

ax[0,0].plot(x, cos_x, label='cos x', c='r', ls='--')
ax[0,1].plot(x, sin_x, label='sin x')
ax[1,0].plot(x, cos_x, label='cos x', c='g', ls='--')
ax[1,1].plot(x, sin_x, label='sin x', c='g')

for axis in ax.flatten():
    axis.legend()
    axis.grid()

plt.tight_layout # Makes things neater
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10,3),
                      sharex='col', sharey='row') # Share axis

ax[0,0].plot(x, cos_x, label='cos x', c='r', ls='--')
ax[0,1].plot(x, sin_x, label='sin x')
ax[1,0].plot(x, cos_x, label='cos x', c='g', ls='--')
ax[1,1].plot(x, sin_x, label='sin x', c='g')

for axis in ax.flatten():
    axis.legend()
    axis.grid()

plt.tight_layout # Makes things neater
```
